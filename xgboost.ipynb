```python
import numpy as np
import pandas as pd
```

```python
#read data
train_df = pd.read_csv('feature_engineer_dataset/train2.csv', header=None)
test_df = pd.read_csv('feature_engineer_dataset/test2.csv', header = None)
```

```python
train_df
```

```python
col_name=['PassengerId', 'Pclass','Sex','SibSp','Parch','Embarked','Title','FamilySize','AgeRange','FareRange','AgeClassBand', 'Survived']
```

```python
train_df.columns = col_name
```

```python
train_df
```

```python
label=train_df['Survived']
train_data= train_df.drop(['PassengerId','Survived'], axis=1)
```

```python
train_data
```

```python
test_df.columns = col_name[:-1]
```

```python

test_data= test_df.drop(['PassengerId'], axis=1)
test_data
```

```python
import xgboost as xgb
```

```python
from xgboost.sklearn import XGBClassifier
from sklearn import cross_validation, metrics   #Additional scklearn functions
from sklearn.grid_search import GridSearchCV   #Perforing grid search
```

```python
#Tune max_depth and min_child_weight
param_test1 = {
 'max_depth':range(3,15,2),
 'min_child_weight':range(1,8,2)
}
```

```python

gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
 param_grid = param_test1, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
```

```python
gsearch1.fit(train_data, label)
```

```python
gsearch1.grid_scores_, gsearch1.best_params_, gsearch1.best_score_
```

```python
#tune gamma

param_test3 = {
 'gamma':[i/10.0 for i in range(0,10)]
}
gsearch3 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=7, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27), 
 param_grid = param_test3, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch3.fit(train_data, label)
gsearch3.grid_scores_, gsearch3.best_params_, gsearch3.best_score_
```

```python
#tune subsample and colsample_bytree
param_test4 = {
 'subsample':[i/10.0 for i in range(6,10)],
 'colsample_bytree':[i/10.0 for i in range(6,10)]
}
gsearch4 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.15, n_estimators=140, max_depth=5,
 min_child_weight=7, gamma=0.9, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27), 
 param_grid = param_test4, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch4.fit(train_data, label)
gsearch4.grid_scores_, gsearch4.best_params_, gsearch4.best_score_
```

```python
#Tuning Regularization Parameters: alpha
param_test6 = {
 'reg_alpha':[1e-5, 1e-2, 0.1,0.5, 1,2,100]
}
gsearch6 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.15, n_estimators=140, max_depth=5,
 min_child_weight=7, gamma=0.9, subsample=0.9, colsample_bytree=0.9,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27), 
 param_grid = param_test6, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch6.fit(train_data, label)
gsearch6.grid_scores_, gsearch6.best_params_, gsearch6.best_score_
```

```python
#Tuning Regularization Parameters: lambda
param_test7 = {
 'reg_lambda':[1e-5, 1e-2, 0.1, 1, 100]
  #  'reg_alpha':[0.5,1,1.5,2]
}
gsearch7 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.15, n_estimators=140, max_depth=5,
 min_child_weight=7, gamma=0.9, subsample=0.9, colsample_bytree=0.9,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27), 
 param_grid = param_test7, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch7.fit(train_data, label)
gsearch7.grid_scores_, gsearch7.best_params_, gsearch7.best_score_
```

```python
#Learning Rate
param_test8 = {
 'learning_rate':[1e-3, 1e-2,0.05, 0.1, 0.15, 0.2, 1]

}
gsearch8 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=7, gamma=0.9, subsample=0.9, colsample_bytree=0.9,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1,seed=27, reg_alpha=1e-05), 
 param_grid = param_test8, scoring='roc_auc',n_jobs=4,iid=False, cv=5)
gsearch8.fit(train_data, label)
gsearch8.grid_scores_, gsearch8.best_params_, gsearch8.best_score_
```

```python
xgb3 = XGBClassifier(
 learning_rate =0.15,
 n_estimators=1000,
 max_depth=5,
 min_child_weight=7,
 gamma=0.9,
 subsample=0.9,
 colsample_bytree=0.9,
 reg_alpha=1e-05,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)
```

```python
xgdmat = xgb.DMatrix(train_data, label)
```

```python
#build xgboost model
our_params = { 'eta':0.15,
 'n_estimators':1000,
 'max_depth':5,
 'min_child_weight':7,
 'gamma':0.9,
 'subsample':0.9,
 'colsample_bytree':0.9,
 'reg_alpha':1e-05,
 'objective':'binary:logistic',
 'nthread':4,
 'scale_pos_weight':1,
 'seed':27} 

# Grid Search CV optimized settings
cv_xgb = xgb.cv(params = our_params, dtrain = xgdmat, num_boost_round = 3000, nfold = 5,
                metrics = ['error'], # Make sure you enter metrics inside a list or you may encounter issues!
                early_stopping_rounds = 100) # Look for early stopping that minimizes error
```

```python
cv_xgb.tail(5)
```

```python
#easly stop after 17 round
final_gb = xgb.train(our_params, xgdmat, num_boost_round = 17)
```

```python
#important features analysis
%matplotlib inline
import seaborn as sns
sns.set(font_scale = 1.5)
```

```python

xgb.plot_importance(final_gb)
```

```python
#prediction on test data
testdmat = xgb.DMatrix(test_data)
```

```python
from sklearn.metrics import accuracy_score
y_pred = final_gb.predict(testdmat) # Predict using our testdmat
#y_pred
```

```python
y_pred[y_pred > 0.5] = 1
y_pred[y_pred <= 0.5] = 0
#y_pred
```

```python
#get true labels for test dataset
y_test=pd.read_csv('data/gender_submission.csv')
```

```python
y_test=y_test.drop(['PassengerId'], axis=1)
```

```python
#calculate accuracy
accuracy_score(y_pred, y_test), 1-accuracy_score(y_pred, y_test)
```
